```python
import pandas as pd
```

```python
data = pd.read_csv("Updated Spreadsheet - Sheet1 (1).csv")
```

```python
data.head()
```

```python
def find_transitions(labels, segment_duration=2):
    labels_non_na = labels.dropna()
    max_time = int(labels_non_na[-2])
    result = [0] * (max_time // segment_duration + 1)
    for i in range(3, len(labels_non_na), 3):
        result[int(labels_non_na[i] / segment_duration)] = 1
    return result           
```

```python
data["transitions"] = data.apply(find_transitions, axis=1)
```

```python
data.to_csv("transitions.csv", index=False)
```
